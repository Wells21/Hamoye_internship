Question 18

```python
data_new = data.drop(columns = ['date', 'lights'])
```

```python
data_new.head()
```

```python
y = data_new[['Appliances']]
X = data_new.drop(columns = ['Appliances'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
names = scaler.get_feature_names_out()
```

```python
names
```

```python
X_train = pd.DataFrame(X_train, columns=names)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test = pd.DataFrame(X_test, columns=names)
```

```python
X_test.head()
```

```python
X_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_preds = model.predict(X_train)
```

```python
mse = round(mean_absolute_error(y_train, y_train_preds), 3)
print(mse)
```

Question 19

```python
rmse = round(np.sqrt(mean_squared_error(y_train, y_train_preds)), 3)
print(rmse)
```

Question 20

```python
y_test_preds = model.predict(X_test)
```

```python
mse = round(mean_absolute_error(y_test, y_test_preds), 3)
print(mse)
```

Question 21

```python
rmse = round(np.sqrt(mean_squared_error(y_test, y_test_preds)), 3)
print(rmse)
```
